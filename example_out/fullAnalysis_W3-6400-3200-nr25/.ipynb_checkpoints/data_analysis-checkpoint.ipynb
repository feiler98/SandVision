# Analysis of the output Excel sheet

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
file_name = "fullAnalysis_W3-6400-3200-nr25__eval.xlsx"
```

```python
df = pd.read_excel(file_name).sort_values("Unnamed: 0").set_index("Unnamed: 0")
df.head(5)
```

```python
df["diff_angle"] = df["angle_h_abs_chamber"] - df["angle_h_abs_sand"] 
```

```python
list_time_progress = [int(x.split("_")[-1]) for x in df.index]
list_time_progress_zero = [x-min(list_time_progress) for x in list_time_progress]
```

```python
df["rel_time"] = list_time_progress_zero
```

```python
plt.scatter(df["rel_time"], df["diff_angle"], label="diff angle")
plt.scatter(df["rel_time"], df["angle_h_abs_sand"], label="sand angle")
plt.scatter(df["rel_time"], df["angle_h_abs_chamber"], label="chamber angle")
plt.xlabel("time [ms]")
plt.ylabel("angle [deg]")
plt.legend()
plt.show()
```
